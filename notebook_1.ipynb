```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average values
average_values = df.groupby('Gender2').agg(
    average_net_worth=('net worth', 'mean'),
    average_annual_salary=('annual Salary', 'mean')
).reset_index()

# Melt the dataframe for easier plotting with seaborn
average_values_melted = average_values.melt(id_vars='Gender2', var_name='Metrics', value_name='Average Value')

# Set the style and context
sns.set(style='whitegrid', context='talk')

# Create the bar plot
plt.figure(figsize=(14, 8))
barplot = sns.barplot(data=average_values_melted, x='Gender2', y='Average Value', hue='Metrics', palette='viridis')

# Add title and labels
plt.title('Comparison of Average Net Worth and Average Annual Salary by Gender', fontsize=18, weight='bold')
plt.xlabel('Gender', fontsize=14, weight='bold')
plt.ylabel('Average Value (in dollars)', fontsize=14, weight='bold')

# Update legend
plt.legend(title='Metrics', title_fontsize='13', fontsize='12', loc='upper right')

# Add value labels to each bar
for container in barplot.containers:
    barplot.bar_label(container, fmt='%.2f', label_type='edge', padding=3, fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()

```

```python
def age_group_creations(row):
    if row['age'] < 30:
        return '20-30'
    elif row['age'] < 40:
        return '30-40'
    elif row['age'] < 50:
        return '40-50'
    elif row['age'] < 60:
        return '50-60'
    else:
        return '60-70'


df['age_group'] = df.apply(age_group_creations, axis=1)


df[['age', 'age_group']].head()
```

# Feature Engineering / Machine Learning Models

```python
X = df[['age','annual Salary','net worth']]
y = df[['car purchase amount']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.2)
```

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

```python
X_train = scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler,'scaler.pkl')
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

def modelresults(predictions):
    print(f'Mean absolute error is {mean_absolute_error(y_test, predictions):.4f}')
    print(f'Mean squared error is {np.sqrt(mean_squared_error(y_test, predictions)):.4f}')
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
predictions_linear_regression = lr.predict(X_test)
```

```python
modelresults(predictions_linear_regression)
```

   - The **RMSE** of 725.2174 suggests that, considering the squared differences, the average error is higher. However, it's still a small fraction of the mean value.
